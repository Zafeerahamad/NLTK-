```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

In the above cell we import the stemmer 

**Stemmer** means it converts word into base form of words 

```python
tokenizer=RegexpTokenizer(r'\w+')
en_stopwords=set(stopwords.words('english'))
ps=PorterStemmer()
```

```python
def getCleanedText(text):
    text=text.lower()
    
    # tokenize
    tokens=tokenizer.tokenize(text)
    new_tokens=[token for token in tokens if (token not in en_stopwords)]
    stemmed_tokens=[ps.stem(tokens) for tokens in new_tokens]
    clean_text=" ".join(stemmed_tokens)
    return clean_text
```

```python
c_text= [getCleanedText(i) for i in list_resume]
```

## user define NAive bayes classifier

```python
def prior_probability(train_x,train_y,label):
    total_examples=train_x.shape[0]
    class_examples=np.sum(train_y==label)
    return (class_examples/float(total_examples))
```

```python
def cond_prob(train_x,train_y,feature_column,feature_value,label):
    x_filtered=train_x[train_y==label]   # only those rows of training data where y_train =class label
    numerator= np.sum(x_filtered[:,feature_column]==feature_value)  # along all filtered rows where value of featur column == feature value
    denomenator=np.sum(train_y==label)    
    return (numerator/float(denomenator))
```

### Now we are going to Compute the Posterior probability or the prediction

```python
def predict(x_train,y_train,xtest):
    '''x_test is a single testing point ,n features'''
    n_features=x_train.shape[1]
    post_prob=[]
    classes=np.unique(y_train)
    for label in classes:
        likelihood=1.0
        for f in range(n_features):
            cond=cond_prob(x_train,y_train,f,xtest[f],label)
            likelihood*=cond
        prior=prior_probability(x_train,y_train,label)
        post=likelihood*prior
        post_prob.append(post)
    pred=np.argmax(post_prob)
    return pred
```
